```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
df.head(10)
```

```python
df['Sex'].value_counts().plot(kind='bar')
plt.title('Sex distribution')
plt.ylabel('Count')
plt.show()
```

```python
number_of_survived = df['Survived'].value_counts().to_dict()
plt.pie(number_of_survived.values(),labels=['Dead','Survived'])
plt.title('Number of people survived and dead in the ship')
plt.show()
```

```python
males = df[df['Sex'] == 'male']
males_survived = males[males['Survived'] == 1]
number_of_males_survived = len(males_survived)
males_dead = males[males['Survived'] == 0]
number_of_males_dead = len(males_dead)
plt.pie([number_of_males_dead,number_of_males_survived],labels=['males dead','males survived'])
plt.title('Males dead and survived')
plt.show()
```

```python
females = df[df['Sex'] == 'female']
females_survived = females[females['Survived'] == 1]
number_of_females_survived = len(females_survived)
females_dead = females[females['Survived'] == 0]
number_of_females_dead = len(females_dead)
plt.pie([number_of_females_dead,number_of_females_survived],labels=['females dead','females survived'])
plt.title('females dead and survived')
plt.show()
```

```python
oldest_person = max(df['Age'])
print(f'oldest person is {oldest_person} years old')
youngest_person = min(df['Age'])
print(f'youngest person is {youngest_person} years old')
average_age = df['Age'].mean()
print(f'average of ages : {int(average_age)}')
```

```python
data = df[['Age', 'Fare']].dropna()
ages = data['Age']
fares = data['Fare']
plt.scatter(ages,fares)
plt.xlabel('age')
plt.ylabel('fare')
plt.title('relationship between fare and age')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('student_dataset_10000_rows.csv')
df.head(20)
```

```python
df.info()
```

```python
placed_status_numbers = df['placement_status'].value_counts().to_dict()
plt.bar(placed_status_numbers.keys(),placed_status_numbers.values(),color=['tab:blue','tab:red'])
plt.title('Number of students placed and failed')
plt.ylabel('Number of students')
plt.show()
```

```python
placed_students = df[df['placement_status'] == 'Placed']
failed_students = df[df['placement_status'] == 'Not Placed']
average_placed_st_h = placed_students['study_hours'].mean()
average_failed_st_h = failed_students['study_hours'].mean()
average_placed_sl_h = placed_students['sleep_hours'].mean()
average_failed_sl_h = failed_students['sleep_hours'].mean()
average_placed_iu = placed_students['internet_usage'].mean()
average_failed_iu = failed_students['internet_usage'].mean()
plt.bar(['study ','study  ','sleep ','sleep  ','net usage ','net usage  '],[average_placed_st_h,average_failed_st_h,average_placed_sl_h,average_failed_sl_h,average_placed_iu,average_failed_iu],color=['tab:blue','tab:red','tab:blue','tab:red','tab:blue','tab:red'])
plt.title('red:failed , blue:placed')
plt.ylabel('average')
plt.xlabel('field')
plt.show()
```

```python
placed_average_assignments_completed = placed_students['assignments_completed'].mean()
failed_average_assignments_completed = failed_students['assignments_completed'].mean()
plt.bar(['assignments_completed ','assignments_completed'],[placed_average_assignments_completed,failed_average_assignments_completed],color=['tab:blue','tab:red'])
plt.ylabel('average')
plt.xlabel('field')
plt.show()
```

```python
placed_average_previous_score = placed_students['previous_score'].mean()
failed_average_previous_score = failed_students['previous_score'].mean()
plt.bar(['placed student ','failed students'],[placed_average_previous_score,failed_average_previous_score],color=['tab:blue','tab:red'])
plt.xlabel('previous score')
plt.ylabel('average')
plt.show()
```
